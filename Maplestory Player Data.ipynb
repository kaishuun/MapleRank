```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq 
import pandas as pd
import numpy as np
```

```python
def scraper(url):
    #opens connection, grabs html, closes connection
    client = uReq(url)
    page_html = client.read()
    client.close()


    page_soup = soup(page_html,"html.parser")
    table = page_soup.find("table")

    #finds and gets the job data
    jobs = [job.get("alt") for job in table.findAll("img",{"class":"job"})]

    #finds and gets the name data
    i = 0
    names = []
    tddata = table.findAll("td")

    while 2+ i*6 <= len(table.findAll("td")):
        names.append(tddata[2+i*6].text)
        i += 1

    #finds and gets the world
    worlds = [world.get("class")[1] for world in table.findAll("a")]

    levels = [int(level.text.split()[0]) for level in table.findAll("td",{"class":"level-move"})]

    data = pd.DataFrame()
    data["Character Name"] = names
    data["World"] = worlds
    data["Job"] = jobs
    data["Level/Move"] = levels
    return data
```

```python
url = 'http://maplestory.nexon.net/rankings/overall-ranking/legendary?rebootIndex=0&pageIndex='

i = 0
df = pd.DataFrame()
detect = True

while detect:
    link = url + str(i)
    df = df.append(scraper(link))
    if not (200 <= df["Level/Move"]).all():
        detect = False
    i += 5
```
