# Data Cleaning + Transforming
Due to the sheer amount of Characters being over Level 200 we make slight modifications to the data we:
* Consider Characters $>=$ level 205
* Remove null values from the collection process
* Add on Class labels (Warrior, Magician, Thief, Pirate, Archers)
* Add on Job Type labels (Explorers, Cygnus Knights, Heroes, Resistance, Nova, Sengoku, Flora, Others)


```r
df = read_csv("Maplestory Rank data.csv",col_types = cols(Rank = col_double(), `Character Name` = col_character(), World = col_factor(), Job = col_factor(), `Level/Move` = col_double()))

#filters to level 205 and above only
df <- df %>% filter(`Level/Move` >= 205, !is.na(Job))
head(df)
```

```r
#adds Job type column (Sengoku, Explorer, ect)
Explorers <- c("Warrior","Magician","Bowman","Pathfinder","Thief","Dual Blade","Pirate","Cannoneer","Jett")
Cygnus_Knights <- c("Dawn Warrior","Blaze Wizard","Wind Archer","Night Walker","Thunder Breaker","Mihile")
Heroes <- c("Aran","Evan","Mercedes","Phantom","Luminous","Shade")
Resistance <- c("Blaster","Battle Mage","Wild Hunter","Mechanic","Demon Avenger","Demon Slayer","Xenon")
Nova <- c("Kaiser","Cadena","Angelic Buster")
Sengoku <- c("Kanna","Hayato")
Flora <- c("Illium","Ark")
Others <- c("Zero","Beast Tamer","Kinesis","Hoyoung")

jobs <- list(Explorers,Heroes,Cygnus_Knights,Resistance,Nova,Sengoku,Flora,Others)
jobtitles <- c("Explorers","Heroes","Cygnus Knights","Resistance","Nova","Sengoku","Flora","Others")

result = data.frame()
for( i in 1:length(jobs)){
    temp <- df %>% filter(Job %in% jobs[[i]]) %>% mutate(`Job Type` = jobtitles[i])
    result <- result %>% rbind(temp)
}
df <- result

#adds Class type (Magician,Warrior,ect)
magician <- c("Battle Mage","Beast Tamer","Blaze Wizard","Evan","Kanna","Luminous","Magician","Kinesis","Illium")
thief <- c("Dual Blade","Night Walker","Phantom","Thief","Xenon","Cadena","Hoyoung")
warrior <- c("Aran","Dawn Warrior","Demon Avenger","Demon Slayer","Hayato","Kaiser","Mihile","Warrior","Zero","Blaster")
bowman <- c("Bowman","Wild Hunter","Wind Archer","Mercedes","Pathfinder")
pirate <- c("Angelic Buster","Cannoneer","Jett","Mechanic","Pirate","Shade","Thunder Breaker","Ark")

classes <- list(magician, thief, warrior, bowman, pirate)
classtitles <- c("Magician","Thief","warrior","Bowman","Pirate")

result = data.frame()
for( i in 1:length(classes)){
    temp <- df %>% filter(Job %in% classes[[i]]) %>% mutate(`Class` = classtitles[i])
    result <- result %>% rbind(temp)
}
df <- result
head(df)
```

# Summary Statistics
## World Statistics
Since the reboot server's play style drastically differs from normal world, so we take a look at the distribution in each seperate world

```r
#max level/character 
nonrebootlv <- df %>% group_by(World,Job) %>% summarize(`Max Lv` = max(`Level/Move`)) %>% ggplot(aes(x=reorder(factor(Job),`Max Lv`), y = `Max Lv`)) 
nonrebootlv + geom_bar(stat = "identity", aes(fill = Job), show.legend = FALSE)  +facet_grid(~World) +coord_flip(ylim=c(210, 276)) + xlab("Job") + ggtitle("Maximum Level for Each Job in Each World")
```

```r
#distribution of players between worlds
df %>% ggplot(aes(x=World)) + geom_bar(aes(fill = World)) + ylab("Population over 205") + ggtitle("Number of Players Per World")
```

Here we can see the sheer force of the reboot server, which has a higher population than almost all of the servers combined. Reboot has the most characters at the level cap (275) with over 50% of Jobs at the level cap. In this comparison there exists one Mihile at level cap in Reboot, while the max in other servers is just 275.

```r
#distribution of players over 205
df %>% ggplot(aes(x=`Level/Move`)) + geom_histogram(bins = 14) + coord_cartesian(xlim = c(205,275)) + facet_wrap(~World) +  ggtitle("Level Distribution per World")
```

```r
df %>% filter(`Level/Move` >= 250) %>% ggplot(aes(x=`Level/Move`)) + geom_histogram(bins = 14) + coord_cartesian(xlim = c(250,275)) + facet_wrap(~World) + ggtitle("Level Distribution for Level 250+ per World")
```

The distribution of levels after 205 peaks around level 210, followed by an exponential decrease until level 275 which is the current maximum. It reflects pretty well as leveling gets increasingly more difficult the higher you go. All of the worlds follow this same distribution of levels.

```r
#distribution for each character in each world
df %>% ggplot(aes(x = Job, y = `Level/Move`)) + geom_boxplot() + coord_flip() + facet_grid(~World) + ggtitle("Level Distrbution for Each Job per World")
```

At a big surprise, Mechanics has a larger IQR in reboot than in normal servers; and, at no surprise Kannas have the highest median level in reboot due to popularity in Meso farming.

## Comparisons of Classes and Job Type
We now dive specifically into the big question of "What class should I main?" by looking directly at popularity data

```r
#Job distribution per world
jobdist <- df %>% group_by(Job) %>% mutate(temp = 1,total = sum(temp)) %>% ungroup() %>%group_by(World)
jobdist %>% ggplot(aes(x= reorder(Job,total))) + geom_bar(aes(fill = Job), show.legend = FALSE) + coord_flip() + facet_grid(~World) +theme(axis.text.x = element_text(angle = 270))+ xlab("Job") + ggtitle("Popularity of Each Job")
```

```r
#no kanna player distrbution per world
jobdist %>% filter(Job != "Kanna") %>% ggplot(aes(x= reorder(Job,total))) + geom_bar(aes(fill = Job), show.legend = FALSE) + coord_flip() + facet_grid(~World) + theme(axis.text.x = element_text(angle = 270))+xlab("Job") + ggtitle("Popularity of Each Job Minus Kanna")
```

Here in these population chart, we confirm the popularity of Meso farming with a Kanna in the reboot server. With reboot heavily replying on Mesos, Kanna's popularity almost triples the second most popular class. Even in normal servers, Kannas are still at the top of the chart.

One interesting note is that players in Bera tend to make comparibly more Dual Blades than in other servers


```r
#What Job types do players like? (thief, warrior, ect)
jobdist %>% ggplot(aes(x= reorder(Job,total))) + geom_bar(aes(fill = Job), show.legend = FALSE) + coord_flip() + facet_grid(~Class) + theme(axis.text.x = element_text(angle = 270))+xlab("Job") + ggtitle("Popularity of Each Job Within Each Class")
```

```r
#What Job grouping do players like? (sengoku, explorer, ect)
jobdist %>% ggplot(aes(x= reorder(Job,total))) + geom_bar(aes(fill = Job), show.legend = FALSE) + coord_flip() + facet_grid(~`Job Type`) + theme(axis.text.x = element_text(angle = 270))+xlab("Job") + ggtitle("Popularity of Each Job within Each Job Type")
```

We now finally answer the questions of: "What job should I make that's in {X class}?" and "What {Job Type} job is good?

At a surprise, despite being one of the oldest job types, explorers still dominate in popularity for almost all classes. As well, heroes are evenly distributed with Evan being the least popular out of all Heroes.

Also comparing the new Flora classes, it's seen that more people definetly enjoy Ark's playstyle more than Illium's.
